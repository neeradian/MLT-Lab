```python
import numpy as np
import pandas as pd
```

```python
# Load dataset
data = pd.DataFrame(data=pd.read_csv('../datasets/Data2.csv'))
```

```python
# Extract concepts and target
concepts = np.array(data.iloc[:, 0:-1])
print(concepts)
```

```python
target = np.array(data.iloc[:, -1])
print(target)
```

```python
def learn(concepts, target):
    # Initialize specific hypothesis
    specific_h = concepts[0].copy()
    print("Initialization of specific_h and general_h")
    print(specific_h)

    # Initialize general hypothesis
    general_h = [["?" for i in range(len(specific_h))]
                 for i in range(len(specific_h))]
    print(general_h)

    # Learning steps
    for i, h in enumerate(concepts):

        # If positive example
        if target[i] == "Yes":
            for x in range(len(specific_h)):
                if h[x] != specific_h[x]:
                    specific_h[x] = '?'
                    general_h[x][x] = '?'

        # If negative example
        if target[i] == "No":
            for x in range(len(specific_h)):
                if h[x] != specific_h[x]:
                    general_h[x][x] = specific_h[x]
                else:
                    general_h[x][x] = '?'

        print(f"Steps of Candidate Elimination Algorithm {i + 1}")
        print(specific_h)
        print(general_h)

    # Remove overly general hypotheses
    indices = [
        i for i, val in enumerate(general_h)
        if val == ['?' for _ in range(len(specific_h))]
    ]

    for i in indices:
        general_h.remove(['?' for _ in range(len(specific_h))])

    return specific_h, general_h
```

```python
# Run learning algorithm
s_final, g_final = learn(concepts, target)
```

```python
print("Final Specific_h:")
print(s_final)
```
