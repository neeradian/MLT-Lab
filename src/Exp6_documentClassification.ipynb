```python
import pandas as pd
```

```python
msg = pd.read_csv('../datasets/Data6.csv', names=['message', 'label'])
print("Total Instance of Dataset: ", msg.shape[0])
msg['labelnum'] = msg.label.map({'pos': 1, 'neg': 0})
```

```python
x = msg.message
y = msg.labelnum
```

```python
from sklearn.model_selection import train_test_split
Xtrain, Xtest, ytrain, ytest = train_test_split(x, y)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

count_v = CountVectorizer()
Xtrain_dm = count_v.fit_transform(Xtrain)
Xtest_dm = count_v.transform(Xtest)
```

```python
df = pd.DataFrame(Xtrain_dm.toarray(), columns=count_v.get_feature_names_out())
print(df[0:5])
```

```python
from sklearn.naive_bayes import MultinomialNB

clf = MultinomialNB()
clf.fit(Xtrain_dm, ytrain)
pred = clf.predict(Xtest_dm)
# pred
```

```python
for doc, p in zip(Xtest, pred):
    p = 'pos' if p == 1 else 'neg'
    print("%s -> %s" % (doc, p))
    # print(f"{doc} -> {p}")
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, recall_score

print("Accuracy Metrics: \n")
print('Accuracy: ', accuracy_score(ytest, pred))
print('Recall: ', recall_score(ytest, pred))
print('Prediction:', precision_score(ytest, pred))
print('Confusion Matrix: \n', confusion_matrix(ytest, pred))
```
