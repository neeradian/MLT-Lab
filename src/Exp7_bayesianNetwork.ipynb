```python
import pandas as pd
```

```python
# Load dataset
data = pd.read_csv("../datasets/Data7.csv")
heart_disease = pd.DataFrame(data)
print(heart_disease)
```

```python
# Import Bayesian Network model
from pgmpy.models import DiscreteBayesianNetwork
```

```python
# Define the structure of the Bayesian Network
model = DiscreteBayesianNetwork([
    ('Age', 'Lifestyle'),
    ('Gender', 'Lifestyle'),
    ('Family', 'heartdisease'),
    ('Lifestyle', 'diet'),
    ('diet', 'cholestrol'),
    ('cholestrol', 'heartdisease')
])
```

```python
# Parameter learning using Maximum Likelihood Estimation
from pgmpy.estimators import MaximumLikelihoodEstimator
model.fit(heart_disease, estimator=MaximumLikelihoodEstimator)
```

```python
# Inference
from pgmpy.inference import VariableElimination
HeartDisease_infer = VariableElimination(model)
```

```python
# User instructions
print('For age Enter { SuperSenior Citizen: 0, Senior Citizen: 1, MiddleAged: 2, Youth: 3, Teen: 4 }')
print('For Gender Enter { Male: 0, Female: 1 }')
print('For Family History Enter { Yes: 1, No: 0 }')
print('For diet Enter { High: 0, Medium: 1 }')
print('For Lifestyle Enter { Athlete: 0, Active: 1, Moderate: 2, Sedentary: 3 }')
print('For cholesterol Enter { High: 0, Borderline: 1, Normal: 2 }')
```

```python
# Query the Bayesian Network
q = HeartDisease_infer.query(
    variables=['heartdisease'],
    evidence={
        'Age': int(input('Enter age: ')),               #1
        'Gender': int(input('Enter Gender: ')),         #1
        'Family': int(input('Enter Family history: ')), #0
        'diet': int(input('Enter diet: ')),             #1
        'Lifestyle': int(input('Enter Lifestyle: ')),   #0
        'cholestrol': int(input('Enter cholestrol: '))  #1
    }
)
```
